```python
import pandas as pd
data = pd.read_csv("grafana_2020.csv", header = 0, sep=",",
                  names = ["Time", "Power", "Date", "Month", "Day", "Year"])
del data["Month"]
del data["Day"]
x = data
x
```

```python
import numpy as np
def max_value(file_name, date, Power):
    pd.read_csv(file_name, header = 0, sep=",")
    data = (x.set_index(date).groupby(level=0)["Power"]
            .agg([('Average Power',np.average),('Maximum Power',np.max),('Minimum Power',np.min),("Total Energy Consumed",np.sum)]))
    return data
print((max_value("grafana_2020.csv", x["Date"], x["Power"])))
```
